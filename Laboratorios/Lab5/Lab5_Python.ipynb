```python
rech1 = pd.read_stata(r'../../data/Endes/rech1.dta') 
```

```python
rech1.drop('index', axis = 1, inplace = True)

# inplace = True to replace change in the same dataset 
```

```python
rech1['HVIDX'] = rech1['HVIDX'].apply(str)
```

## Load rech0

```python
rech0 = pd.read_stata('../../data/Endes/rech0.dta') 
```

```python
rech0.info()
```

```python
rech0.columns = map(str.lower, rech0.columns)
```

```python
rech0['hv025'].unique()
```

```python
variables = ["hhid","hv005", "hv009", "hv025", "hv024", "longitudx", "latitudy"]
```

```python
rech0 = rech0[variables]
```

```python
rech0
```

```python
# Recode rural area
rech0.hv025.replace((1, 2), (0, 1), inplace=True)
```

```python
rech0
```

```python
# load Domestic Violence module

rech84 = pd.read_stata('../../data/Endes/rec84dv.dta') 
```

```python
rech84.info
```

```python
rech84['caseid']
```

```python
rech84['caseid']
```

```python
rech84['caseid'] = rech84['caseid'].str[:9] + rech84['caseid'].str[9:12].str.strip()
```

```python
rech84
```

```python
index = np.where( rech84.columns.str.contains('d101'))[0]

names = list(rech84.columns[index])
```

```python
names[1]
```

```python
for i in names:
    rech84[f'{i}'] = rech84[f'{i}'].replace(9,np.nan)
    rech84[f'{i}'] = rech84[f'{i}'].replace(99,np.nan)

```

```python
rech84
```

```python
outer_merged = pd.merge(rech1, rech0, how="outer", on=["hhid"])
```

```python
outer_merged
```

```python
inner_merged = pd.merge(rech1, rech0, how="inner", on=["hhid"])
inner_merged
```

```python
left_merged = pd.merge(rech1, rech0, how="left", on=["hhid"])
left_merged 
```

```python
right_merged = pd.merge(rech1, rech0, how="right", on=["hhid"])
right_merged 
```

```python
base_completa = pd.merge(left_merged, rech84, how="left", on=["caseid"])
```
