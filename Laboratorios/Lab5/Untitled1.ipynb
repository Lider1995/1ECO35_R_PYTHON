```r
relative_path = '../../data/Endes/504-Modulo64'

dataset <- c("RECH0","RECH1")

for (j in dataset){
    
    x <- read.dbf(glue('{relative_path}/{j}.dbf'))
    assign( glue('{j}'), x) 
    
    write.dta( x , glue('../../data/Endes/{j}.dta'))
    
}
```
