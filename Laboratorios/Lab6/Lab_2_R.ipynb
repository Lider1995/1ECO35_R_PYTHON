```r
panel <- read.dta13("../../data/ENAHO/panel_2016_2018.dta", encoding = "latin1")
```

```r
panel
```

```r
panel <- panel %>% rename("year_16" = "a<d1>o_16", "year_17" = "a<d1>o_17", "year_18" = "a<d1>o_18", "year_19" = "a<d1>o_19",
                         "year_20" = "a<d1>o_20", "hogar" = "hogar_16")
```

```r
panel
```

```r
names <- names(panel)
```

## 1.0 Substract letters an d number, but not special characters 

```r
filter_list <- c()

for (i in 1:length(names)){
        w <-  gsub('[^-A-Za-z0-9]','', names[i])
        filter_list[i] <- w
        print(w)    
}
```

## Drop special character '_'

```r
filter_list <- c()

for (i in 1:length(names)){
        w <-  gsub('[/_]','', names[i])
        filter_list[i] <- w
        print(w)    
}
```

## 2.0 Substract just letters (capital or lower)

```r
filter_list <- c()

for (i in 1:length(names)){
        w <-  gsub("[^a-z+]",'', names[i])
        filter_list[i] <- w
        print(w)    
}
```
