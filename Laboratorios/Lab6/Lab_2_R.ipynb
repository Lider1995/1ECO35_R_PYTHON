## 1.0 Substract letters and number, but not special characters 

```r
filter_list <- c()

for (i in 1:length(names)){
        w <-  gsub('[^-A-Za-z0-9]','', names[i])
        filter_list[i] <- w
        print(w)    
}
```

## Drop special character '_'

```r
filter_list <- c()

for (i in 1:length(names)){
        w <-  gsub('[/_]','', names[i])
        filter_list[i] <- w
        print(w)    
}
```

## 2.0 Substract just letters (capital or lower)

```r
filter_list <- c()

for (i in 1:length(names)){
        w <-  gsub("[^A-Za-z]",'', names[i])
        filter_list[i] <- w
        print(w)    
}
```

```r
filter_list <- c()

for (i in 1:length(names)){
        w <-  gsub("[^0-9]",'', names[i])
        filter_list[i] <- w
        print(w)    
}
```
