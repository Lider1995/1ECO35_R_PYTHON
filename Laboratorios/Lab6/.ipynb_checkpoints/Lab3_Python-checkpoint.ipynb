```python
import pandas as pd
import numpy as np
import chardet # to get string character format 
import re
```

```python
panel = pd.read_stata("../../data/ENAHO/panel_2016_2018.dta")
```

```python
panel.columns = map(str.lower, panel.columns)  # any capittal letter to lower
```

```python
panel.rename(columns = {'año_16':'year_16', 'año_17':'year_17', 'año_18':'year_18', 'año_19':'year_19',
                       'año_20':'year_20'}, inplace = True)
```

```python
# rename hogar_16 to build a identifier per household 

panel.rename(columns = {'hogar_16':'hogar'}, inplace = True)
```

```python
list(panel.columns)
```

```python
filter_list = []

for i in range(0,len(list(panel.columns))):

        w = re.sub('[^-A-Z-a-z]+','', list(panel.columns)[i])
        filter_list.append(w)
        
    
```

```python
new_list = list(dict.fromkeys(filter_list))
```

```python
new_list = new_list[7:57] # keep interest variables 
new_list.append('year')
```

```python
new_list
```

```python
panel = panel.set_index(['conglome','vivienda','hogar'])
```

```python
panel['idx'] = panel.groupby(['conglome','vivienda','hogar']).cumcount()
```

```python
panel
```

```python
panel.stack()
```

```python
panel.pivot(index=['conglome','vivienda','hogar'], columns= 'year_16' ,values=new_list)
```

```python
panel
```

```python
tmp = []
for var in new_list:
    panel['tmp_idx'] = var + '_' + panel.idx.astype(str)
    tmp.append(panel.pivot(index=['conglome','vivienda','hogar'],columns='tmp_idx',values=var))

reshape = pd.concat(tmp,axis=1)
```
