## Filtro de base de datos

### 1. Quedarse con los hogares Urbanos, asimismo considerar los hogares que completaron la encuesta.

```python
#solo me voy a quedar con aquella información que se ha recolectado de la zona urbana

urban = enapres2020_1.loc[ enapres2020_1.AREA == 'URBANO', : ]
urban
```

```python
#Me voy a quedar con aquellas personas que contestan la encuesta completa - complet survey (RESFIN)

urban_c = urban.loc[ urban.RESFIN == 'Completa', : ]
urban_c
```

### 2. Verificar si existe duplicados de hogares

```python
#Hay duplicados por hogar
#Se pone (keep=False) para que me muestre las variables que tienen duplicado y sus duplicados (mantiene ambas) 

urban_c[ urban_c.loc[:, ['CCDD' ,'CCPP' , 'CCDI' ,'CONGLOMERADO' , 'NSELV', 'VIVIENDA', 'HOGAR']].duplicated(keep=False) ]
```

```python
#las variables que tienen duplicados
#Lo bueno es que solo son 10 errores de 40 000
urban_c[ urban_c.loc[:, ['CCDD' ,'CCPP' , 'CCDI' ,'CONGLOMERADO' , 'NSELV', 'VIVIENDA', 'HOGAR']].duplicated() ]
```

### 3. Si existiese, filtrar la base de datos para no considerar los duplicados.

```python
#Borrar los 5 duplicados de los hogares, para que no me afecte el análisis.

urban_no_dp = urban_c[ ~urban_c.loc[:, ['CCDD' ,'CCPP' , 'CCDI' ,'CONGLOMERADO' , 'NSELV', 'VIVIENDA', 'HOGAR']].duplicated() ].copy()
urban_no_dp
```

## Crear Dummies

### 1. Crear una Dummy si la zona donde se ubica el hogar instaló cámaras de seguridad. (variable P122A_1). Tenga cuidado con los missing values si los hubiera.

```python
urban_no_dp["P122A_1"]
```

```python
urban_no_dp["P122A_1"].isnull().sum()
```

```python
# Aplicar esta sentencia al crear sus variables pues P122A_1 presenta missing values. De esta forma se crea variables dummies con presencia de missing values
urban_no_dp['Dummy_1'] = np.where(urban_no_dp["P122A_1"] == "Si", 1, np.where(urban_no_dp["P122A_1"].notnull(), 0, np.nan))
```

```python
urban_no_dp['Dummy_1'] 
```

```python
urban_no_dp['Dummy_1'].unique()
```

```python
urban_no_dp["P184A_14"]
```

```python
urban_no_dp["P184A_14"].isnull().sum()
```

```python
# create a dummy 

urban_no_dp['Dummy_2'] = np.where(urban_no_dp["P184A_14"]== "¿Conexión a Internet?", 1, 0)

# Estructura  where:  np.where( condition, verdadero: coloca 1, falso: coloca 0)
```

```python
urban_no_dp["P107A"].unique()
```

```python
urban_no_dp["P107A"].isnull().sum()
```

```python
# Aplicar esta sentencia al crear sus variables pues P107A presenta missing values. De esta forma se crea variables dummies con presencia de missing values: 

urban_no_dp['Dummy_3'] = np.where(urban_no_dp['P107A'] == "Título de Propiedad", 1, np.where(urban_no_dp['P107A'].notnull(), 0, np.nan))

# Cuando es falso la sentencia no coloca 0 sino que colocará cero si el valor es diferente a un missing value 
```

## Group by

### 1. Obtener el total de hogares por distrito, el total de hogares que cuentan con internet por distrito, finalmente el porcentaje de hogares que cuentan con internet.

```python
df = urban_no_dp.groupby( [ 'CCDD' ,'CCPP' , 'CCDI' ] , as_index = False ).agg( total_hogares_distrito = ('HOGAR', np.size), total_hogares_internet = ("Dummy_2", np.sum) )
```

```python
df
```

```python
df["Porcentaje_hogares_con_internet"] = df["total_hogares_internet"] / df["total_hogares_distrito"] 
df
```

### 2. Obtener una base de datos a nivel departamentos con la siguiente información: total de hogares, total de hogares cuyo vecindario instaló cámaras de seguridad y el porcentaje hogares cuyos vecindarios cuentan con cámaras de seguridad. Base_1

```python
Base__1 = urban_no_dp.groupby( [ 'CCDD'] , as_index = False ).agg( total_hogares_departamento = ('HOGAR', np.size), total_hogares_camara = ("Dummy_1", np.sum) )
Base__1
```

```python
Base__1["Porcentaje_hogares_con_camara"] = (Base__1["total_hogares_camara"] / Base__1["total_hogares_departamento"])*100
Base__1
```

### 3. Crear una base de datos a nivel departamentos con la siguiente información: total de hogares que cuentan con títulos de propiedad y porcentaje de hogares que cuentan con títulos de propiedad. Base_2

```python
Base__2 = urban_no_dp.groupby( [ 'CCDD'] , as_index = False ).agg( total_hogares_departamento = ('HOGAR', np.size), total_hogares_titulo = ("Dummy_3", np.sum) )
Base__2
```

```python
Base__2["Porcentaje_hogares_con_titulo"] = Base__2["total_hogares_titulo"] / Base__2["total_hogares_departamento"] 
Base__2
```

## Merge

### 1. Hacer un merge entre las bases de datos Base_1 y Base_2

```python
Base__1 = Base__1.merge( Base__2 , how = 'left', on = ["CCDD",'total_hogares_departamento'], indicator = False)
Base__1
```

```python
Clean_data=Base__1
Clean_data
```

### 2. Finalmente guardar la base de datos en el formato SPSS. Llamar a la base de datos (Clean_data)

```python
# save in spss format 

import pyreadstat

pyreadstat.write_sav(Clean_data,  r"../data/enapres_2020_ch_100/736-Modulo1618/test1.sav")
```
