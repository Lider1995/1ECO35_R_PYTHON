## 2. Filter data

```r
enapres2020_1 <- enapres2020_1 %>% filter(AREA == 1 & RESFIN == 1)
```

```r
# duplicates obervations

enapres2020_1 %>% group_by(CCDD ,CCPP , CCDI ,CONGLOMERADO , NSELV, VIVIENDA, HOGAR) %>% 
                  mutate(duplicates = n()) %>% filter(duplicates >1) %>%
                select(CCDD ,CCPP , CCDI ,CONGLOMERADO , NSELV, VIVIENDA, HOGAR) 
```

```r
df_urban_no_dpl <- enapres2020_1 %>% distinct(CCDD ,CCPP , CCDI ,CONGLOMERADO , NSELV, VIVIENDA, HOGAR, .keep_all = TRUE)
```

```r
dim(df_urban_no_dpl)
```

## 3.0 Dummies section 

```r
# checking missing values 

unique(P122A_1)
unique(P184A_14)
unique(P107A)
```

```r
df_urban_no_dpl <- df_urban_no_dpl %>%  mutate(Dummy_1 = ifelse(P122A_1 == 1 ,  1 , ifelse(!is.na(P122A_1),0, NA) ) ) %>%
                                            mutate(Dummy_2 = ifelse(P184A_14 == 1 ,  1 , 0 ) )  %>% 
                                            mutate(Dummy_3 = ifelse(P107A== 1 ,  1 , ifelse(!is.na(P107A),0, NA) ) )
```

## 3.1 Households access to internet

```r
base1 <- df_urban_no_dpl %>% group_by(CCDD ,CCPP , CCDI) %>% 
                    summarise(total_hogares_distrito = n() , hogar_internet = sum(Dummy_2)) %>%
                    mutate( acceso_internet =  round((hogar_internet/total_hogares_distrito)*100, 2))
```

```r
head(base1)
```

## 3.2 Households and Security system

```r
Base_1 <- df_urban_no_dpl %>% group_by(CCDD) %>% 
                    summarise(total_hogares = n() , hogar_seguridad = sum(!is.na(Dummy_1))) %>%
                    mutate( seguridad_porcentaje =  round((hogar_seguridad/total_hogares)*100, 2))
```

```r
head(Base_1)
```

## 3.3 Households and title deeds

```r
Base_2 <- df_urban_no_dpl %>% group_by(CCDD) %>% 
                    summarise(total_hogares = n() , hogar_titulo = sum(!is.na(Dummy_3)) ) %>%
                    mutate( titulo_propiedad_porcentaje =  round((hogar_titulo/total_hogares)*100, 2))
```

```r
head(Base_2)
```

```r
base_total <- merge( Base_1, Base_2, by = c("CCDD","total_hogares"), all.x = TRUE)
```

```r
write_sav(base_total,"../../data/Clean_data.sav")  # save spps format using "haven" library
```
