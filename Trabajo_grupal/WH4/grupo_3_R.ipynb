## Filtro de base de datos
1. Quedarse con los hogares Urbanos, asimismo considerar los hogares que completaron la encuesta.
2. Verificar si existe duplicados de hogares
3. Si existiese, filtrar la base de datos para no considerar los duplicados

```r
#etiquetas de los valores de la variable RESFIN
enapres2020$RESFIN %>% attr('labels')
```

```r
#etiquetas de la variable AREA
enapres2020$AREA %>% attr('labels')
```

```r
#1. Quedarse con los hogares Urbanos, asimismo considerar los hogares que completaron la encuesta.



ENA <-enapres2020[(enapres2020$AREA %in% c(1)) & (enapres2020$RESFIN %in% c(1)),]
ENA

```

```r
#2. Verificar si existe duplicados de hogares
ENA %>% group_by(CCDD ,CCPP , CCDI ,CONGLOMERADO , NSELV, VIVIENDA, HOGAR) %>% 
                  mutate(duplicates = n()) %>% filter(duplicates >1) %>%
                select(CCDD ,CCPP , CCDI ,CONGLOMERADO , NSELV, VIVIENDA, HOGAR) 

# First group by unique household identifier
# mutate() create variables
# filter duplicates 
# select unique household identifier
```

```r
#3. Si existiese, filtrar la base de datos para no considerar los duplicados

ENA_ndup <- ENA %>% distinct(CCDD ,CCPP , CCDI ,CONGLOMERADO , NSELV, VIVIENDA, HOGAR, .keep_all = TRUE)
ENA_ndup
```

## Crear Dummies

3. Crear una Dummy si el hogar cuenta con titulo de propiedad (variable P107A)

```r
# Para ver missing values en P122A_1
sum(is.na(ENA_ndup$P122A_1))
# Otra opcion
unique(ENA_ndup$P122A_1)

```

## Group by
1. Obtener el total de hogares por distrito, el total de hogares que cuentan con internet por distrito, finalmente el porcentaje de hogares que cuentan con internet.
2. Obtener una base de datos a nivel departamentos con la siguiente información: total de hogares, total de hogares cuyo vecindario instaló cámaras de seguridad y el porcentaje hogares cuyos vecindarios cuentan con cámaras de seguridad. Base_1
3. Crear una base de datos a nivel departamentos con la siguiente información: total de hogares que cuentan con títulos de propiedad y porcentaje de hogares que cuentan con títulos de propiedad. Base_2

```r
#1. Obtener el total de hogares por distrito, el total de hogares que cuentan con internet por distrito,
#   finalmente el porcentaje de hogares que cuentan con internet

# Total hogares por distrito
distrito <- ENA_ndup %>% group_by(NOMBREDI) %>% 
                tally %>% 
               rename(total_hogar_distrito = n)

distrito
```

```r
# Para la segunda parte de la pregunta 1
#Primero: filtramos internet
internet <- ENA_ndup %>% filter(dummy_internet %in% c(1))
df1 <- as.data.frame (internet)
df1
```

```r
# Segundo: total de hogares que cuentan con internet por distrito
internet_dist <- df1 %>% group_by(NOMBREDI, dummy_internet) %>% 
                tally %>% 
               rename(internet_dist = n)

internet_dist
```

```r
# Para la tercera parte de la pregunta 1.
# El porcentaje de hogares que cuentan con internet por distrito.
porcent <- df1 %>%
  group_by(NOMBREDI, dummy_internet) %>%
  summarise(internet_distrito = n()) %>%
  mutate(porcent_inter = round(100*(internet_distrito / sum(ENA_ndup$TOT_HOGAR)), 3)) 
                                                  
porcent 
```

```r
# 2.- Obtener una base de datos a nivel departamentos con la siguiente información: 
# total de hogares, total de hogares cuyo vecindario instaló cámaras de seguridad y
#el porcentaje hogares cuyos vecindarios cuentan con cámaras de seguridad. Base_1

# Para la primera parte de la pregunta 2
# total de hogares por dpto. 

total_hog_dpto <- ENA_ndup %>% group_by(NOMBREDD) %>% tally %>% rename(total_hogares_dpto = n)
total_hog_dpto
```

```r
# Para la segunda parte de la pregunta 2

#Primero: filtramos los hogares cuyo vecindario instaló cámaras de seguridad
camara <- ENA_ndup %>% filter(dummy_camara %in% c(1))

df2 <- as.data.frame (camara)
df2
```

```r
# Segundo: total de hogares cuyo vecindario instaló cámaras de seguridad por dpto.
camara_dpto <- df2 %>% group_by(NOMBREDD) %>% 
                tally %>% 
               rename(camara_dpto = n)

camara_dpto
```

```r
# Para la tercera parte de la pregunta 2.
# El porcentaje de hogares cuyo vecindario instaló cámaras de seguridad por dpto.
porcent <- df2 %>%
  group_by(NOMBREDD) %>%
  summarise(camara_dpto = n()) %>%
  mutate(porcent_camara = round(100*(camara_dpto / sum(ENA_ndup$TOT_HOGAR)), 3)) 
                                                  
porcent
```

```r
# Crear una base de datos con toda la informacion anterior: Base_1
NOMBREDD <- total_hog_dpto$NOMBREDD
total_hogares <- total_hog_dpto$total_hogares_dpto
camara <-camara_dpto$camara_dpto
porcent_camara <- porcent$porcent_camara


df01 <- cbind(NOMBREDD, total_hogares, camara, porcent_camara)
base_1 <- as.data.frame(df01)
base_1
```

```r
# 3. Crear una base de datos a nivel departamentos con la siguiente información: 
# total de hogares que cuentan con títulos de propiedad y 
# porcentaje de hogares que cuentan con títulos de propiedad. Base_2

# Para la primera parte de la pregunta 3
#filtramos titulo
titulo <- ENA_ndup %>% filter(dummy_titulo %in% c(1))
df3 <- as.data.frame (titulo)
df3
```

```r
# Porcentaje de hogares que cuentan con títulos de propiedad
porcentaje <- df3 %>%
  group_by(NOMBREDD) %>%
  summarise(titulo_dpto = n()) %>%
  mutate(porcent_titulo = round(100*(titulo_dpto / sum(ENA_ndup$TOT_HOGAR)), 3)) 
                                                  
porcentaje
```

```r
#uniendo para base de datos
base_2 <- as.data.frame(porcentaje)
base_2
```

## Merge
5. Hacer un merge entre las bases de datos Base_1 y Base_2


```r
base_total <- merge(  base_1, base_2, by = c("NOMBREDD"), all.x = TRUE)


# base_1 (x), base_2 (y)
```
