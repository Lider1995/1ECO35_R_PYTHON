```r
# Crear una funcion que me permita extraer una columna segun el nombre de la variable y dos filas.
var_2rows <- function(df,var){
    # Seleccionando 2 filas
 df <- head(df,2)
  # Seleccionando la variable
 df <- df[,var]
  
  return(df)
}
```
