## 7.0 Drop variables 

```r
junin_data <- junin_data[! names(junin_data) %in% c('finished_instr_total', 'finished_instr_men' , 'finished_instr_women')]
```

## 8.0 Percentage variables 

```r
attach(junin_data)

junin_data  <-  junin_data %>% mutate(total_population = peruvian_men + peruvian_women+foreign_men+foreign_women,  
                                     '%women_not_read_write' = (nlee_mujer+nesc_mujer)*100/total_population,
                                     '%men_not_read_write' = (nlee_hombre+nesc_hombre)*100/total_population,
                                     '%natives' = (natives)*100/total_population)

```

```r
head(junin_data)
```

## 9.0 New dataset  

```r
distritos <- c('CIUDAD DEL CERRO', 'JAUJA', 'ACOLLA', 'SAN GERÓNIMO', 'TARMA', 'OROYA' , 'CONCEPCIÓN')


newdata <- junin_data[District %in% distritos,]
```

```r
head(newdata)
```

```r
newdata <- newdata %>% filter( (newdata['4_6_years_total'] > 50) & (newdata['6_14_years_total'] > 70) ) %>%
                       arrange(total_population,natives )

```

```r
head(newdata)
```

```r
base_final <- newdata[,c('District','comunidad','%women_not_read_write','%men_not_read_write','%natives','total_population')]
```

### 10.0 Function section

```r
function_1 <- function(Base,c1,c2,r1,r2){
    
    X = Base[r1:r2,c1:c2]
    
    return(X)
}

function_1(newdata, 1,31,4,5)
```

```r
function_2 <- function(Base,var,r1,r2){
    
    X = Base[c(r1,r2),var]
    
    return(X)
}

function_2(newdata, 'comunidad',3,6)
```
