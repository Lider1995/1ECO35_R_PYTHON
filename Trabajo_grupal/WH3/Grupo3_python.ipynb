8. Crear columnas con las siguiente información: el % de mujeres que no escriben ni leen, % de varones que no escriben ni leen y % de nativos respecto al total de la población.

```python
df['totalpoblacion'] = df['peruvian_men'] + df['peruvian_women'] + df['foreign_men'] + df['foreign_women']
df[f'%mujeres_noescriben_nileen'] = (df['nlee_mujer'] + df['nesc_mujer'])/ df['totalpoblacion']
df[f'%hombres_noescriben_nileen'] = (df['nlee_hombre'] + df['nesc_hombre']) / df['totalpoblacion']
df[f'%nativos'] = df['natives'] / df['totalpoblacion']
df
```
