Crear una función que me permita extraer columnas un rango de columnas y de filas. Crear una función que me permite extraer una columna según el nombre de la variable y dos filas.

```python
def extract_1(df,x,y,m,n):
    result = df.loc[m:n,(x,y)]
    return result
        
```

```python
extract_1(junin_data, 'District', 'Place', 1, 3)
```

```python
def extract_2(df,x,m,n):
    result = df.loc[m:n,x]
    return result
        
```

```python
extract_2(junin_data, 'District', 1, 3)
```
