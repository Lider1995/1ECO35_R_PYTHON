7. Borrar las varaibles finished_instr_total, finished_instr_men y finished_instr_women

8. Crear columnas con las siguiente información: el % de mujeres que no escriben ni leen, % de varones que no escriben ni leen y % de nativos respecto al total de la población.

```python
B["Population"] = B["peruvian_men"] + B["peruvian_women"] + B["foreign_men"] + B["foreign_women"]
B
```

```python
B["WN_EN_L"] = ((B["nlee_mujer"] + B["nesc_mujer"])/B["Population"])*100
B
```

```python
B["MN_EN_L"] =  ((B["nlee_hombre"] + B["nesc_hombre"])/B["Population"])*100
B
```

```python
B["Nativos"] = (B["natives"]/B["Population"])*100
B
```

9.Crear una base de datos con la siguiente información:

a. Quedarse con la información de los distritos de Ciudad del Cerro, Jauja, Acolla, San Gerónimo, Tarma, Oroya y Concepción

```python
D = B[(B["District"] == "CIUDAD DEL CERRO")| (B["District"] == "JAUJA") | (B["District"] == "ACOLLA") | (B["District"] == "SAN GERÓNIMO") | (B["District"] == "TARMA") | (B["District"] == "OROYA") | (B["District"] == "CONCEPCIÓN")]
D
```

b. Ciudades con una cantidad de niños (4-6 años) mayor a 50 y de (6-14 años) > 70

```python
nin= D[(D["4_6_years_total"]> 50) & (D["6_14_years_total"]> 70)]
nin
```

c. Ordenar las variables según el total de la población y cantidad de población nativa

```python
E = nin.sort_values(["Population","natives"], ascending = [True,True])
E
```
