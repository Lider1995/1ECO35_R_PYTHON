## 7.0 Drop variables 

```python
junin_data.drop(['finished_instr_total', 'finished_instr_men' , 'finished_instr_women'], axis=1 , inplace = True)
```

## 8.0 Percentage variables 

```python
junin_data['total_population'] = junin_data['peruvian_men']+junin_data['peruvian_women']+junin_data['foreign_men']+junin_data['foreign_women']
```

```python
junin_data['%women_not_read_write'] = ((junin_data['nlee_mujer']+junin_data['nesc_mujer'])/junin_data['total_population'])*100

junin_data['%men_not_read_write'] = ((junin_data['nlee_hombre']+junin_data['nesc_hombre'])/junin_data['total_population'])*100


junin_data['%natives'] = (junin_data['natives']/junin_data['total_population'])*100
```

## 9.0 New dataset   

```python
distritos = ['CIUDAD DEL CERRO', 'JAUJA', 'ACOLLA', 'SAN GERÓNIMO', 'TARMA', 'OROYA' , 'CONCEPCIÓN']

new_data = junin_data[junin_data["District"].isin(distritos)]
```

```python
new_data
```

```python
new_data.sort_values(["total_population","natives"], inplace = True)
```

### 10.0 Function section

```python
def function_1(Base,c1,c2,r1,r2):
    X = Base.iloc[r1:r2,c1:c2]
    
    return X
```

```python
function_1(Base = new_data,c1 = 1 ,c2 = 30,r1 = 3, r2 = 5)
```

```python
def function_2(Base,var,r1,r2):
    
    X = Base[[var]].iloc[[r1,r2],:]
    
    return X
```

```python
function_2(new_data,'comunidad',2,5)
```
