# Operaciones con matrices

7- Obtener el estimador de la varianza del termino de perturbación

```python
import random
```

```python
random.seed(456)

x1 = np.random.rand(20) # uniform distribution  [0,1]
x2 = np.random.rand(20) # uniform distribution [0,1]
x3 = np.random.rand(20) # uniform distribution [0,1]
x4 = np.random.rand(20) # uniform distribution [0,1]
e = np.random.normal(0,1,20) # normal distribution mean = 0 and sd = 1

# Poblacional regression (Data Generating Process GDP)

Y = 20 + 5*x1 + 9*x2 + 6*x3 + 4*x4 + e
```

```python
X = np.column_stack((np.ones(20),x1,x2,x3,x4))
X
```

```python
trX=np.transpose(X)
print(trX)
```

```python
X_X = trX @ X
X_X
```

```python
invX_X = np.linalg.inv(X_X)
invX_X
```

```python
W = invX_X @ trX
W
```

```python
Z_betas = W @ Y
print(Z_betas)
```

```python
#Procedimiento directo
Betas = np.linalg.inv(X_X) @ (trX @ Y )
print(Betas)
```

```python
#Tomando en cuenta el modelo estimado (Y_hat=X*beta)
Y_hat = X @ Betas
print(Y_hat)
```

```python
#Para hallar la SRC (tr_e_hat*e_hat). Primero se halla e_hat (e_hat = Y - Y_hat) y su traspuesta

e_hat = Y - Y_hat
e_hat
```

```python
#Sacando la traspuesta de e_hat

tr_e_hat = np.transpose(e_hat)
tr_e_hat
```

```python
#Hallando la SRC ((e_hat)'*e_hat)

SRC = tr_e_hat @  e_hat
SRC
```

```python
#Hallando la SD, teniendo en cuenta que se tiene 20 observaciones con k=5 (4 variables explicativas y la constante). Entonces, n-k=20-5=15
SD = SRC/15
SD
```

# Matriz generadora de residuos

\begin{equation}
 M = I - X(X'X)^{-1}X'
\end{equation}

9 - Matriz generador de residuos

```python
#Operando primero las matrices: X(X'X)^{-1} = X*invX_X

G = X @ invX_X
G
```

```python
#Entonces, M ahora es: M = I - G*X'; pero H = G*trX. Entonces, M = I - H

H = G @ trX
H
```

```python
#Solo falta definir una matriz identidad de (20x20)

I = np.eye(20)
print(I)
```

```python
#Finalmente, M = I - H
M = I-H
M
```
