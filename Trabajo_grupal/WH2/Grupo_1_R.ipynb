- Use un loop while para mostrar el nombre de las personas ingresantes a la carrera de quimia.

```r
notes <- read.csv("../../data/notas.csv")
notes
```

```r
notes['Condicion'] <- notes$notas

i <- 1

nrows = dim(notes)[1]

while (i <= nrows){
    
     if (notes$notas[i] >= 11){
        notes$Condicion[i] = 'Aprobo'
        }
    else if (notes$notas[i] < 11){
        notes$Condicion[i] = 'No aprobo'
    }
    i <- i + 1 
}
notes
```
