# 1. IF STATEMENT

Un archivo .csv llamado notas está en la carpeta data. Esta base de datos posee información de algunos postulantes universitarios a la universidad. Use if statement para crear una nueva columna e indicar si el el alumno ingreso o no (nota minima aprobatoria es 11). Luego crear un columna para colocar el nombre completo de la carrera a la que postulo el alumno (a001 : Economia, a002: quimica, a003: filosofia).

```python
import pandas as pd

postulantes = pd.read_csv(r"..\..\data\notas.csv")

postulantes
```

```python
i=0
nrows = postulantes.notas.count()

while i < nrows:
    if postulantes['notas'][i] < 11:
        postulantes.loc[[i],["Estado"]] = "no ingresó"
    else:
        postulantes.loc[[i],["Estado"]] = "ingresó" 
    
    i += 1 

postulantes
```

```python
i=0
nrows = postulantes.notas.count()

while i < nrows:
    if postulantes['carrera'][i] == "a001":
        postulantes.loc[[i],["nombre_carrera"]] = "ECONOMÍA"
    elif postulantes['carrera'][i] == "a002":
        postulantes.loc[[i],["nombre_carrera"]] = "QUÍMICA"
    elif  postulantes['carrera'][i] == "a003":
        postulantes.loc[[i],["nombre_carrera"]] = "FILOSOFÍA"
    else:
        postulantes.loc[[i],["nombre_carrera"]] = "-"
    
    i += 1 
    
postulantes
```

### Simule un proceso generador de datos con 10 mil observaciones como se explicó en la primera práctica dirigida. Use un loop para obtener los coeficientes estimados con los siguientes tamaños de muestra (10, 50, 100,200,500,1000,5000). Guardar los coeficientes estimados en un Dataframe. Nombre las columnas del dataframe con los coeficientes estimados (i.e b1, b2, b3 ...) ; y las filas, con el tamaño de muestra (n=10 , n= 50, ... ).

```python
import random 
import numpy as np
import pandas as pd

random.seed(2000)

x1 = np.random.rand(10000) # uniform distribution  [0,1]
x2 = np.random.rand(10000) # uniform distribution [0,1]
x3 = np.random.rand(10000) # uniform distribution [0,1]
x4 = np.random.rand(10000) # uniform distribution [0,1]
e = np.random.normal(0,1,10000) # normal distribution mean = 0 and sd = 1

# Poblacional regression (Data Generating Process GDP)


Y = 1 + 0.3*x1 + 0.75*x2 - 4.3*x3 + 3.6*x4 + e
```

```python
X = np.column_stack((np.ones(10000),x1,x2,x3,x4))
X
```

```python
#Betas estimados de toda la muestra (n=10000)
beta = np.linalg.inv(X.T @ X) @ ((X.T) @ Y )
print(beta)
```

```python
num = (10,50,100,200,500,1000,5000)
betas = np.zeros((7,5))

for i in np.arange(0,len(num)):
    Xn = X[1:num[i],:]
    betas[i,:] = np.linalg.inv(Xn.T @ Xn) @ ((Xn.T) @ Y[1:num[i]] )
    #print(Xn)
    #print(a[i])
    #print(betaN)
    
    
num1 = np.asarray(num)
df = pd.DataFrame(betas, columns = ['b1','b2','b3','b4','b5'])
df['N'] = num1
#df1 = pd.DataFrame((df.N,df.b1,df.b2,df.b3,df.b4,df.b5).transpose())
print(df)
```

### (Ejercicio solo en Python) Usar la base de datos places result para crear un DataFrame con la siguiente información: nombre de la institución pública, dirección (formatted_address), situación de actividad (business_status) y las coordenadas correspondientes a northeast y southwest.

```python
#pip install tqdm
```

```python
from tqdm import tqdm   #measure time
import pickle 
```

```python
dictionary_places = open( r"..\..\data\places_result", "rb")
places_result = pickle.load( dictionary_places )
```

```python
places_result
```

```python
places_result.keys()
```

```python
# Lists
institutions = []
direccion = []
status = []

# define all the results
results = places_result[ 'results' ]

# loop para guardar cada uno de los elementos
for row in tqdm( range( 0 , len( results ) ) ):
    
    # nombre de las institucion
    name = results[ row ]['name']
    
    # Direccion
    direcc = results[ row ]['formatted_address']
    
    # Status de la institucion
    stat = results[ row ]['business_status']
    
    # Save results
    status.append( stat )
    direccion.append( direcc )
    institutions.append( name )

    # Diccionario
final_result = { 'Institution': institutions, 'Direccion' : direccion, 'Status' : status }
# Dataframe
df_result = pd.DataFrame( final_result )
df_result
```
