```python
#Un archivo .csv llamado notas está en la carpeta data. 
#Esta base de datos posee información de algunos postulantes universitarios a la universidad. Use if statement para crear una nueva columna e indicar si el el alumno ingreso o no (nota minima aprobatoria es 11). Luego crear un columna para colocar el nombre completo de la carrera a la que postulo el alumno 
#(a001 : Economia, a002: quimica, a003: filosofia).

import pandas as pd
notas_1 = pd.read_csv(r"..\..\data\notas.csv")
notas_1
```

```python

# Creando nueva columna con descripción ("ingreso" , "no ingreso")
n=0
newcol = notas_1.notas.count()

while n < newcol:
    if notas_1['notas'][n] < 11:
        notas_1.loc[[n],["Estado"]] = "no ingreso"
    else:
        notas_1.loc[[n],["Estado"]] = "ingreso" 
    
    n += 1 

notas_1
```

```python
#clasificación de postulantes según carrera.
n=0
nrows = notas_1.notas.count()

while n < nrows:
    if notas_1['carrera'][n] == "a001":
        notas_1.loc[[n],["Carrera"]] = "Economia"
    elif notas_1['carrera'][n] == "a002":
        notas_1.loc[[n],["Carrera"]] = "Quimica"
    elif  notas_1['carrera'][n] == "a003":
        notas_1.loc[[n],["Carrera"]] = "Filosofia"
    else:
        notas_1.loc[[n],["Carrera"]] = "NA"
    
    n += 1 
    
notas_1
```

# LOOPS


Simule un proceso generador de datos con 10 mil observaciones como se explicó en la primera práctica dirigida. Use un loop para obtener los coeficientes estimados con los siguientes tamaños de muestra (10, 50, 100,200,500,1000,5000). Guardar los coeficientes estimados en un Dataframe. Nombre las columnas del dataframe con los coeficientes estimados (i.e b1, b2, b3 ...) ; y las filas, con el tamaño de muestra (n=10 , n= 50, ... ).

```python
import random 
import numpy as np

random.seed(10000)
muestra= [10,50,100,200,500,1000,5000]
```

```python
result = []
for i in range(0,7):
    x1 = np.random.rand(muestra[i]) # uniform distribution  [0,1]
    x2 = np.random.rand(muestra[i])# uniform distribution [0,1]
    x3 = np.random.rand(muestra[i])
    e = np.random.normal(0,1,(muestra[i])) # normal distribution mean = 0 and sd = 1
# Poblacional regression (Data Generating Process GDP)
    M = 0.1 + 4*x1 + 3.2*x2 + 4*x3
    N = np.column_stack((np.ones(muestra[i]), x1,x2))
    beta = np.linalg.inv(N.T @ N) @ ((N.T) @ M )
    result.append(beta)
    print(beta)
```

```python
coef_est = np.vstack(result)
coef_est
```

```python
df2 = pd.DataFrame(coef_est, index= muestra, columns=['b1','b2','b3'])
df2
```

(Ejercicio solo en Python) Usar la base de datos places result para crear un DataFrame con la siguiente información: nombre de la institución pública, dirección (formatted_address), situación de actividad (business_status) y las coordenadas correspondientes a northeast y southwest.

```python
from tqdm import tqdm  
import pickle 
```

```python
dictionary_places = open( r"..\..\data\places_result", "rb")
places_result = pickle.load( dictionary_places )
```

```python
places_result
```

```python
places_result.keys()
```

```python
resultados = places_result[ 'results' ]
```

```python
instituciones = []
direccion = []
actividad = []
northeast = []
southwest = []

# loop para guardar cada uno de los elementos
for row in tqdm( range( 0 , len( resultados ) ) ):
    
    # direccion
    dire = resultados[ row ]['formatted_address']
    
    # estado de la actividad
    act = resultados[ row ]['business_status']
    
    # nombre de la institution
    institution = resultados[ row ]['name']
    
    # southwest
    south = resultados[ row ]['geometry']['viewport']['southwest']
    
    # northwest
    north = resultados[ row ]['geometry']['viewport']['northeast']
    
     # Save results
    instituciones.append( institution )
    direccion.append( dire )
    actividad.append( act )
    northeast.append( north )
    southwest.append( south )
```

```python

resultado_final = { 'Institucion': instituciones, 'Direccion' : direccion, 'Estado' : actividad, 'Direccion_norte' : northeast, 'Direccion_sur' : southwest }

df_final = pd.DataFrame( resultado_final )
df_final
```
