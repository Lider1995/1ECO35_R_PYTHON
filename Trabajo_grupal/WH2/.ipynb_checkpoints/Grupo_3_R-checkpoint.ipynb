```r
#OTRO EJEMPLO
set.seed(123)
# tamaño de muestra = 10
ns <- 10

```

```r
for(i in 1:10000){
        
    x1 <- runif(ns)
    x2 <- runif(ns)
    x3 <- runif(ns)
    e <- rnorm(ns)
    
    #Ejemplo de Y:
    Y = 12 + 3*x1 + 5*x2 + 1*x3 + e
    X <- cbind(matrix(1,ns), x1,x2,x3)
    
    #Estimación de los paramétros mediante OLS
    b_hat1 <- solve( t(X) %*% X) %*% ( t(X) %*% Y )
        }

#resultados
b_hat1
```

4- Use la base de datos production_fish.csv para obtener el porcentaje de consumo de ciertos tipos de producción marina respecto al consumo de anchoveta. Crear esta información solo para los siguiente tipos producción marina: Atun, Jurel, Langostino, Tollo

```r
#production_fish = read.csv("../Data/production_fish.csv" )  
# relative path 

# Usar el siguiente code para la presentacion del HW.
production_fish = read.csv("../../data/production_fish.csv" )
```

```r
production_fish
```

```r
# Ver la lista de variables de la base de datos
names (production_fish)
```

```r
# Definir un vector de produccion de anchoveta
anchoveta <- production_fish$Consumo.Industrial...Anchoveta

# Definir una matriz con los tipos de produccion
atun <-  production_fish$Consumo.Humano...Conservas...AtÃºn

jurel <- production_fish$Consumo.Humano...Conservas...Jurel +
            production_fish$Consumo.Humano...Fresco...Jurel
langostino <-  production_fish$Consumo.Humano...Congelado...Langostino
tollo <-  production_fish$Consumo.Humano...Fresco...Tollo

typeprod_mat <- cbind( anchoveta, atun, jurel, langostino, tollo)

# Visualización de typeprod_mat
# typeprod_mat
```

```r
# Creación de un objeto data frame a partir de la matriz typeprod_mat
typeprod <- as.data.frame( typeprod_mat)

#Visualización de typeprod
typeprod
```

```r
colums <- names(typeprod)
colums
```

```r
#install.packages("glue")
library(glue)
```

```r
glue('{col}_%')
```

```r
for (col in colums){
    typeprod[glue('{col}_%')]   <- (typeprod[glue('{col}')]  * 100) / 
                                    typeprod$anchoveta
    }
```

```r
# Resultados del porcentaje de produccion
typeprod
```

6-Crear una matrix aleatoria de orden 100x10. Crear un función que permita estandariza cada columna; es decir, cada observación de una columna debe restarse por su media y dividirse por la desviación estándar de la columna. Llame a esta función como standarize.

```r
# Creación de la matriz y de orden 100x10
# con sample para generar numeros enteros en modo aleatorio
y <- matrix(sample(1:10000), 100, 10)
y
```

```r
#apply(y, 2, sd)
```

```r
# Creación de la función de estandarización
standarize <- function (y){
    z <- ( y- mean(y) ) / sd(y)
}
```

```r
std_y <- apply(y, 2, standarize)
std_y
```
