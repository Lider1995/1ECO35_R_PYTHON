1. If statement:
Un archivo .csv llamado notas está en la carpeta data. Esta base de datos posee información de algunos postulantes universitarios a la universidad. Use if statement para crear una nueva columna e indicar si el el alumno ingreso o no (nota minima aprobatoria es 11). Luego crear un columna para colocar el nombre completo de la carrera a la que postulo el alumno (a001 : Economia, a002: quimica, a003: filosofia).

```python
import pandas as pd
post_U = pd.read_csv("../data/notas.csv")
post_U
```

```python
i=0
nrows = post_U.notas.count()

while i < nrows:
    if post_U ["notas"][i] > 11:
        post_U.loc[[i],["Estado de admisión"]] = "ADMITIDO"
    else:
        post_U.loc[[i],["Estado de admisión"]] = "NO ADMITIDO" 
    
    i += 1 

post_U
```

```python
i=0
nrows = post_U.notas.count()

while i < nrows:
    if post_U["carrera"][i] == "a001":
        post_U.loc[[i],["Carrera"]] = "ECONOMÍA"
    elif post_U["carrera"][i] == "a002":
        post_U.loc[[i],["Carrera"]] = "QUÍMICA"
    elif  post_U["carrera"][i] == "a003":
        post_U.loc[[i],["Carrera"]] = "FILOSOFÍA"
    else:
        post_U.loc[[i],["Carrera"]] = " "
    
    i += 1 
    
post_U
```

2. loops:
Simule un proceso generador de datos con 10 mil observaciones como se explicó en la primera práctica dirigida. Use un loop para obtener los coeficientes estimados con los siguientes tamaños de muestra (10, 50, 100,200,500,1000,5000). Guardar los coeficientes estimados en un Dataframe. Nombre las columnas del dataframe con los coeficientes estimados (i.e b1, b2, b3 ...) ; y las filas, con el tamaño de muestra (n=10 , n= 50, ... ).

```python
import numpy as np
```

```python
import random 
random.seed(10000)
muestra= [10,50,100,200,500,1000,5000]

result = []
for i in range(0,7):
    x1 = np.random.rand(muestra[i])
    x2 = np.random.rand(muestra[i]) 
    x3 = np.random.rand(muestra[i])
    x4 = np.random.rand(muestra[i])
    e  = np.random.normal(0,1,(muestra[i])) 

    
    Y = 20 + 5*x1 + 9*x2 + 6*x3 + 4*x4 + e
    X = np.column_stack((np.ones(muestra[i]),x1,x2,x3,x4))
    Betas = np.linalg.inv(X.T @ X) @ ((X.T) @ Y )
    result.append(Betas)
    print(Betas)
```

```python
B = np.vstack(result)
print(B)
```

```python
df_result = pd.DataFrame(B,index=[muestra], columns=["B1","B2","B3","B4","B5"] )
print(df_result)
```

Usar la base de datos places result para crear un DataFrame con la siguiente información: nombre de la institución pública, dirección (formatted_address), situación de actividad (business_status) y las coordenadas correspondientes a northeast y southwest.

```python
from tqdm import tqdm   #packages to measure time's loop
import pickle  #to load dictionaries dataset
dictionary_places = open( r"..\data\places_result", "rb") #load dataset
places_result = pickle.load( dictionary_places ) #dictionary format
```

```python
places_result.keys()
```

```python
places_result['results']
```

```python
# Lists
Institutions = []
Direction = []
Business_Status = []
NE = []
SW = []


results = places_result[ 'results' ] #is a list


for row in tqdm( range( 0 , len( results ) ) ):
    
    # INSTITUTIONS
    I = results[ row ]['name']
    
    # DIRECTIONS
    D = results[ row ]['formatted_address']
    
    # BUSINESS STATUS
    BS = results[ row ]['business_status']
    
    
    # COORDINATES
    ##NE
    NE1 = results[ row ]['geometry']['viewport']['northeast']
    
    ##SW
    SW1 = results[ row ]['geometry']['viewport']['southwest']
    
    # Save results
    Institutions.append( I )
    Direction.append( D )
    Business_Status.append( BS )
    NE.append( NE1 )
    SW.append( SW1 )


# Diccionario
Result = { 'Nombre de la Institución': Institutions, 'Dirección' : Direction, 'Situación de Actividad' : Business_Status, 'Coordenada Noreste' : NE,'Coordenada Suroeste' : SW  }

# Dataframe
DF = pd.DataFrame( Result )
DF
```

3. Crear una matrix aleatoria de orden 100x10. Crear un función que permita estandariza cada columna; es decir, cada observación de una columna debe restarse por su media y dividirse por la desviación estándar de la columna. Llame a esta función como standarize.

```python
import numpy as np
matrix = np.arange(0,1000)
Matriz = matrix.reshape(100, 10)  
print(Matriz)
print("Rows:",Matriz.shape[0],"\n", "Columns: ", Matriz.shape[1])
```

```python
M=np.mean(Matriz)
print(M)
```

```python
SD = np.std(Matriz,ddof=1)
print(SD)
```

```python
I_1= np.ones( (100, 10) )
print(I_1)
print("Rows:",I_1.shape[0],"\n", "Columns: ", I_1.shape[1])
```

```python
Media= M*I_1
print(Media)
print("Rows:",Media.shape[0],"\n", "Columns: ", Media.shape[1])
```

```python
A_1 = Matriz - Media
A_1
```

```python
StD= SD*I_1
StD
```

```python
A_2 = A_1 / StD
print(A_2)
```

```python
#Creando la función Standarize

def standarize(Matriz,Media,StD):
    MStandarize = (Matriz - Media) / StD
    
    return MStandarize
result = standarize(Matriz,Media,StD)
print(result)
```
