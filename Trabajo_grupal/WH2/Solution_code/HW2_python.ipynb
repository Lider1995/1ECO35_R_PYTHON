## 2.1 Coefficients loop

```python
random.seed(756)

x1 = np.random.rand(10000) # uniform distribution  [0,1]
x2 = np.random.rand(10000) # uniform distribution [0,1]
x3 = np.random.rand(10000) # uniform distribution [0,1]
x4 = np.random.rand(10000) # uniform distribution [0,1]
e = np.random.normal(0,1,10000) # normal distribution mean = 0 and sd = 1

# Poblacional regression (Data Generating Process GDP)


Y = 1 + 0.8*x1 + 1.2*x2 + 0.5*x3 + 1.5*x4 + e
```

```python
betas = []
sample_size = [10,50,100,200,500,1000,5000]

for i in sample_size:

    X = np.column_stack((np.ones(i),x1[0:i],x2[0:i],x3[0:i],x4[0:i]))
    Y1 = Y[0:i]
    beta = np.linalg.inv(X.T @ X) @ ((X.T) @ Y1 )
    
    betas.append(beta)
```

### Create a dictionary, then convert it to Dataframe

```python
Coeficientes = {'sample size': [10,50,100,200,500,1000,5000], 
     'beta1': np.zeros(7),
      'beta2': np.zeros(7),
    'beta3': np.zeros(7),
    'beta4': np.zeros(7)}
Coeficientes = pd.DataFrame( data = Coeficientes)
Coeficientes
```

### Loop for to fill dataframe with OLS coefficients

```python
for j in range(0,7):
        Coeficientes.iloc[j,1] = betas[j][1]
        Coeficientes.iloc[j,2] = betas[j][2]
        Coeficientes.iloc[j,3] = betas[j][3]
        Coeficientes.iloc[j,4] = betas[j][4]
        
```

## 2.3 Loop - diccionary dataset 

```python
import pickle  # load dictionary dataset 
from tqdm import tqdm   #packages to mesuare time's loop
```

```python
dictionary_places = open( "..\..\..\data\places_result", "rb") # load dataset
places_result = pickle.load( dictionary_places )  # dictionary format 
```

```python
places_result['results'][2]['name']
```

```python
# Lists

latitudes_n = []
longitudes_n = []
latitudes_s = []
longitudes_s = []
institutions = []
address = []
status = []

# define all the results
results = places_result[ 'results' ]   # is a list 
# loop para guardar cada uno de los elementos
for row in tqdm( range( 0 , len( results ) ) ):   #tqdm (time's loop)
    
    # latitude north and latitude south
    lat_n = results[ row ]['geometry']['viewport']['northeast']['lat']
    lat_s = results[ row ]['geometry']['viewport']['southwest']['lat']
    
    # Longitude north and latitude south
    
    lng_n = results[ row ]['geometry']['viewport']['northeast']['lng']
    lng_s = results[ row ]['geometry']['viewport']['southwest']['lng']
    
    # nombre de la institution
    institution = results[ row ]['name']
    
    # address
    Address = results[ row ]['formatted_address']
    
     # status
    Status = results[ row ]['business_status']
    
    # Save results
    latitudes_n.append( lat_n )
    longitudes_n.append( lng_n )
    latitudes_s.append( lat_s )
    longitudes_s.append( lng_s )
    institutions.append( institution )
    address.append(Address)
    status.append( Status )
    
# Diccionario
final_result = { 'Institution': institutions, 'Status': status, 'Address':address ,'North Latitud' : latitudes_n, 'North Longitud' : longitudes_n ,
               'South Latitud' : latitudes_s, 'South Longitud' : longitudes_s}
# Dataframe
df_result = pd.DataFrame( final_result )
df_result
```

```python
def standarize(X):
        
        Z = np.zeros((X.shape[0], X.shape[1]))
        
        for i in range(0,X.shape[1]):
        
            a = np.mean(X[:,i])
            b = np.std(X[:,i])
            Z[:,i] = (X[:,i]-a)/b         
            
        return Z
```
