#### Pregunta 1: If statement:

###### Un archivo .csv llamado notas está en la carpeta data. Esta base de datos posee información de algunos postulantes universitarios a la universidad. Use if statement para crear una nueva columna e indicar si el el alumno ingreso o no (nota minima aprobatoria es 11). Luego crear un columna para colocar el nombre completo de la carrera a la que postulo el alumno (a001 : Economia, a002: quimica, a003: filosofia).

```python
import pandas as pd
notes = pd.read_csv(r"..\..\data\notas.csv")
notes
```

```python
i=0
nrows = notes.notas.count()
while i < nrows:
    
    if notes['notas'][i] >= 11:
        notes.loc[[i],["Condicion"]] = "Aprobo"
    else:
        notes.loc[[i],["Condicion"]] = "No aprobo" 
    
    i += 1 

notes
```

```python
i=0
nrows = notes.notas.count()
while i < nrows:
    
    if notes['carrera'][i] == "a001":
        notes.loc[[i],["nomb_carrera"]] = "Economía"
    elif notes['carrera'][i] == "a002":
        notes.loc[[i],["nomb_carrera"]] = "Química"
    elif  notes['carrera'][i] == "a003":
        notes.loc[[i],["nomb_carrera"]] = "Filosofía"
    else:
        notes.loc[[i],["nomb_carrera"]] = "NA"
    
    i += 1 
    
notes
```
