```r
# REC0111 informaci√≥n b√°sica de la mujer
# REC223132 Informaci√≥n m√≥dulos de salud 
# RE51617 informaci√≥n dela pareja 

# 15-45 a√±os 

range <- c(2015, 2016, 2017, 2018, 2019)

for (year in range) {
      print(year)  

    
    # read REC0111 data
    endes1 <- read.spss(glue('../data/Endes/{year}/REC0111.sav'), to.data.frame=TRUE)
    assign( glue('{year}'), endes1)  

    # Filtering columns
    # selected columns
    endes1_1 <- endes1[, col1 ]
    # generate the year column
    endes1_1$year <- year

    # variables labels
    var_labels <- names(endes1_1)

       
    
    
    # read RE223132 data 
     endes2 <- read.spss(glue('../data/Endes/{year}/RE223132.sav'), to.data.frame=TRUE)
     assign( glue('{year}'), endes2) 
    
    # Filtering columns
    # new pandas with selected columns
    endes2_1 <- endes2[ , col2 ]
    # generate the year column
    endes2_1$year <- year
    
    # variables labels
    var_labels  <- names(endes2_1)
    
    
    
    # read RE516171 data    
     endes3 <- read.spss( glue('../data/Endes/{year}/RE516171.sav'), to.data.frame=TRUE)
     assign( glue('{year}'), endes3) 
    
    # Filtering columns
    # new pandas with selected columns
    endes3_1 <- endes3[ , col3 ]
    # generate the year column
    endes3_1$year <- year
    
    # variables labels
    var_labels  <- names(endes3_1)
    
    
    
    # Merging data
    M <- merge( endes1_1, endes2_1, by = c("CASEID"), all.x = TRUE)
    endes <- merge( M, endes3_1, by = c("CASEID"), all.x = TRUE)
    
    assign( glue('endes_{year}'), endes)
}
```
