La base de datos unidos.dta es una lista de todos los distritos del Perú. Adicionalmente, una variable indica si se aplicó el programa Juntos en el distrito. La base de datos cuenta con la siguiente información: ubigeo, distrito, provincia, region y unidos (Porgrama Juntos applied in the district? Values: 1 (=Yes), 0 (=No).)

```python
Q = []  

for i in range(0,len(list(data.columns))):

        q =  re.sub('[\_]','', list(data.columns)[i])   
        Q.append(q)

# filtrar sin "_"

Q
```

```python
data
```

```python
R = []

for i in range(0,len(list(Q))):

        r =  re.sub(".{2}$",'', list(Q)[i]) 
        R.append(r)
        
# filtrar sin los ultimos dos caracteres (años)

R
```

```python
S = list(dict.fromkeys(R))
S
#Definimos el diccionario: para no tener variables repetidas
```

```python
base = pd.wide_to_long(data, stubnames = S, i = ['numper'] , j = 'year' , sep = '_').reset_index()
base
```

```python
base = base.drop(['nump', 'numpanh', 'numpanh16','numpanh17','numpanh18','numpanh19'], axis=1)
base
```

```python
base2 = base.rename(columns={'numpanh15':'numpanh'})
base2

#Me quedo con las variables más importantes para hacer merge con la base unidos por el ubigeo
```

```python
Base = pd.merge(base2,unidos,on='ubigeo')
Base
```
